```python
g = nx.planted_partition_graph(5, 30, 0.5, 0.05, seed=seed)
pos = nx.spring_layout(g, seed=seed)
blocks = nx.get_node_attributes(g, "block").values()

fig, axes = plt.subplots(nrows=1, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(28)
ax = axes.flatten()

nx.draw(g, ax=ax[0], pos=pos, node_size=200, edge_color="darkgray", node_color="slategray")
nx.draw(g, ax=ax[1], pos=pos, node_size=200, edge_color="darkgray", node_color=list(blocks), vmin=0, vmax=max(blocks), cmap=plt.cm.rainbow)

#ax[0].set_title("A Network", fontsize=25, fontweight="bold")
#ax[1].set_title("After Community Detection", fontsize=25, fontweight="bold")

plt.savefig("../outputs/community_detection.svg", format="svg", dpi=1200)
plt.show()
plt.close()
```
